### DBSCAN

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=250, noise=0.05)
```

```python
data={"data1":x[:,0], "data2":x[:,1]}
```

```python
dataset=pd.DataFrame(data)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.10, min_samples=5)
dataset["Predict"]=db.fit_predict(dataset)
```

```python
sns.scatterplot(x="data1", y="data2",data=dataset,hue="Predict")
plt.show()
```

```python
sns.scatterplot(x="data1", y="data2",data=dataset,hue=y)
plt.show()
```

```python
df=pd.read_csv("Wholesale customers data.csv")
df.head(5)
```

```python
df.info()
```

```python
df.drop(['Channel','Region'],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
array=df.values
```

```python
array
```

```python
array
```

```python
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=6)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```
