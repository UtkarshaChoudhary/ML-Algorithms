```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the data 
data1=pd.read_csv("iris.csv")
```

```python
data1.head(2)
```

```python
#drop the coulumn
data=data1.drop('Unnamed: 0',axis=1)
```

```python
sns.pairplot(data=data)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,10))
sc.dendrogram(sc.linkage(data,method='single',metric='euclidean'))
plt.savefig("Demo.jpg")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac= AgglomerativeClustering(n_clusters=2,linkage='single')
data["Predict"]=ac.fit_predict(data)
```

```python
data.head(2)
```

```python
sns.pairplot(data=data,hue="Predict")
plt.show()
```
