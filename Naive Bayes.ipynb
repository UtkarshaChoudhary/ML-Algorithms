```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
data=pd.read_csv("placed.csv")
```

```python
data.head(3)
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(5,3))
sns.kdeplot(data=data["cgpa"])
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.kdeplot(data=data["iq"])
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa", y="iq", data=data, hue="placement")
plt.show()
```

```python
X= data.iloc[:,:-1]
Y= data["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

## Gaussian Naive Baye's 

```python
gnb= GaussianNB()
gnb.fit(X_train,Y_train)
```

```python
gnb.score(X_test,Y_test)*100, gnb.score(X_train,Y_train)*100
```

```python
plt.figure(figsize=(5,3))
plot_decision_regions(X.to_numpy(),Y.to_numpy(),clf=gnb)
plt.show()
```

### Prediction

```python
gnb.predict([[6.8,123]])
```

## Multinomial Naive Baye's

```python
mnb=MultinomialNB()
mnb.fit(X_train,Y_train)
```

```python
mnb.score(X_test,Y_test)*100 , mnb.score(X_train,Y_train)*100
```

```python
plt.figure(figsize=(5,3))
plot_decision_regions(X.to_numpy(),Y.to_numpy(),clf=mnb)
plt.show()
```

## Bernoulli Naive Baye's

```python
bnb=BernoulliNB()
bnb.fit(X_train,Y_train)
```

```python
bnb.score(X_test,Y_test)*100, bnb.score(X_test,Y_test)*100
```

```python
plt.figure(figsize=(5,3))
plot_decision_regions(X.to_numpy(),Y.to_numpy(),clf=bnb)
plt.show()
```
