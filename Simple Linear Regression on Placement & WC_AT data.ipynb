```python
import pandas as pd
```

```python
data=pd.read_csv("placement.csv")
```

```python
data.head(5)
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa", y="package",data=data)
plt.show()
```

```python
X=data[["cgpa"]]
Y=data["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
#
#6.89	3.26
```

```python
lr.predict([[6.89]])
```

```python
# We are check accuracy score array([2.92962016])
lr.score(X_test,Y_test)*100
# 77.30984312051673 is the accuracy. this is very poor score
```

```python
#Y=m*X+C
lr.coef_    #M value
```

```python
lr.intercept_            #C value
```

```python
#y=mx+c   X  is Independent variable
0.57425647 * 6.89 - 1.02700693745421
```

```python
#exactly we are finding where is regresssion line pass
```

```python
y_pred=lr.predict(X)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa", y="package",data=data)
plt.plot(data["cgpa"],y_pred,c="red")       
plt.legend(["orgdata", "predict_line"])
plt.show()
```

### WC_AT Dataset Linear Regression 

```python
WC_AT=pd.read_csv("WC_AT.csv")
```

```python
WC_AT.head(5)
```

```python
WC_AT.isnull().sum()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Waist", y="AT",data=WC_AT)
plt.show()
```

```python
X=WC_AT[["Waist"]]
Y=WC_AT["AT"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
lr.predict([[74.75]])
```

```python
y_pred=lr.predict(X)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Waist", y="AT",data=WC_AT)
plt.plot(WC_AT["Waist"],y_pred,c="red")       
plt.legend(["orgdata", "predict_line"])
plt.show()
```
